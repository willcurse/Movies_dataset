# DATA CLEANING

```python
#IDENTIFYING OUTLIERS
from scipy import stats
z_score=stats.zscore(data2.select_dtypes(include=['int64','float64']))
outliers=data2[(z_score>3).any(axis=1)|(z_score<-3).any(axis=1)]
```

Outliers are data points that significantly differ from the rest of the data in a dataset. They can be unusually high or low compared to other observations. Here’s a more detailed look at what outliers are and their significance:



```python
#REMOVING THE OUTLIERS

no_outliers=data2[(z_score<3).all(axis=1)&(z_score>-3).all(axis=1)]
no_outliers
```

# Analytics

How Does the Budget of a Movie Influence Its Box Office Performance?

```python
import plotly.express as py
figure = py.scatter(
    data2,
    x='Budget',
    y='Box Office',
    color='Genre',
    title='Budget influence by its box office performance',
    labels={'Budget': 'Budget in $', 'Box Office': 'Box Office in $'}
)
figure.show()
```

Which Directors Have the Highest Average Box Office Percentages?

```python
# Calculate the average box office percentage per director
avg_boxoffice = data2.groupby('Director')['Director Box Office %'].mean().reset_index()

figure = py.bar(
    avg_boxoffice,
    x='Director',
    y='Director Box Office %',
    title='Which director has the highest grossing movie',
    labels={'Director Box Office %': 'Avg Box Office %'}
)

figure.show()
```

How Do Box Office Earnings Vary Across Different Genres?

```python
total_boxoffice=data2.groupby('Genre')['Box Office'].sum().reset_index()
fig=py.bar(
    total_boxoffice,
    x='Genre',
    y='Box Office',
    title='Bar graph for the following',
    labels={'Box Office':'Total'}
    
)
fig.show()
```

```python
data2.columns
```

How Does the Earnings of Movies Change Over the Years?

```python
total_vary_movies=data2.groupby('Release year')['Earnings'].sum().reset_index()
fig=py.scatter(
    total_vary_movies,
    x='Release year',
    y='Earnings',
    labels={'Release year':'year'},
    title='Scatter plot diagram'
)
fig.show()
```

Which Movies Have the Highest Box Office Earnings?

```python
top_10_Movies=data2.nlargest(10,'Box Office')
fig=py.bar(
    top_10_Movies,
    x='Movie',
    y='Box Office',
    title='Top 10 movies',
    labels={'Movies':'Movies names'},
)
fig.show()
```

# KDE plot

```python
numeric_col=['Budget','IMDb score']
fig,axes=plt.subplots(1,2,figsize=(12,5))
for index,col in enumerate(numeric_col):
    sns.kdeplot(data2[col],ax=axes[index])
    axes[index].set_title(f'KDE plot diagram {col}')
plt.tight_layout()
plt.show()
    
```
